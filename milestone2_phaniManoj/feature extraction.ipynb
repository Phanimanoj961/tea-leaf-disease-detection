````python
import numpy as np
from skimage.io import imread
import matplotlib.pyplot as plt

# Load an images
image = imread('test.jpg')
print("Image shape:", image.shape)

# Initialize the feature matrix with the same height and width as the image
feature_matrix = np.zeros((image.shape[0], image.shape[1]))
print("Feature matrix shape:", feature_matrix.shape)

# Calculate the mean pixel value of the channels for each pixel
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        feature_matrix[i, j] = np.mean(image[i, j, :])

# Display the original image and the feature matrix
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image)
plt.subplot(1, 2, 2)
plt.title('Mean Pixel Value Feature Matrix')
plt.imshow(feature_matrix, cmap='gray')
plt.show()
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import exposure, feature
````

````python
## 1. Image Preprocessing

### 1.1 Grayscale Conversion
# Converting an image to grayscale is often the first step in preprocessing. This reduces the complexity by eliminating color information, leaving just intensity.

# ```python
# Load an image
image = cv2.imread('test.jpg')

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the original and grayscale images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.subplot(1, 2, 2)
plt.title('Grayscale Image')
plt.imshow(gray_image, cmap='gray')
plt.show()
````

````python
## 2. Feature Extraction

### 2.1 SIFT (Scale-Invariant Feature Transform)
# SIFT is used to detect and describe local features in images.

# ```python
# Create SIFT detector
sift = cv2.SIFT_create()

# Detect and compute SIFT features
keypoints, descriptors = sift.detectAndCompute(gray_image, None)

# Draw keypoints on the image
sift_image = cv2.drawKeypoints(gray_image, keypoints, None)

# Display the image with keypoints
plt.figure(figsize=(10, 5))
plt.title('SIFT Features')
plt.imshow(sift_image, cmap='gray')
plt.show()
````

````python
### 2.2 HOG (Histogram of Oriented Gradients)
# HOG is used for object detection by counting occurrences of gradient orientation in localized portions of an image.

# ```python
# Compute HOG features
hog_features, hog_image = feature.hog(gray_image, visualize=True)

# Display the HOG image
plt.figure(figsize=(10, 5))
plt.title('HOG Features')
plt.imshow(hog_image, cmap='gray')
plt.show()
````
